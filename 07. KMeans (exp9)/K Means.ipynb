```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import pandas as pd


iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target


# # Define the data points
# data_points = np.array([
#     [1, 2],
#     [1, 4],
#     [1, 0],
#     [4, 2],
#     [4, 4],
#     [4, 0],
#     [7, 2],
#     [7, 4],
#     [7, 0]
# ])


# Set the number of clusters
k = 3

# Create a KMeans instance with fixed random state
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
# Fit the model to the data
kmeans.fit(x)

# Get the cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Print the results
print("Cluster Labels:", labels)
print("Cluster Centers:\n", centers)

# Plotting the results
plt.figure(figsize=(8, 6))
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c=labels, s=100, cmap='viridis', marker='o', label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, alpha=0.75, marker='X', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Define closely grouped data points to form clusters
data_points = np.array([
    [1, 2], [1, 2.1], [1, 1.9],  # Cluster 1
    [4, 4], [4, 4.1], [4, 3.9],  # Cluster 2
    [7, 2], [7, 2.1], [7, 1.9]   # Cluster 3
])

# Set the number of clusters
k = 3

# Fit the K-means model
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data_points)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# New data points for prediction
new_data_points = np.array([
    [1.2, 2],   # Likely to be in cluster 1
    [4.5, 4],   # Likely to be in cluster 2
    [6.5, 2]    # Likely to be in cluster 3
])

# Predict the clusters for the new data points
predicted_clusters = kmeans.predict(new_data_points)

# Print results for original data points
print("Original Data Points and Their Clusters:")
for point, cluster in zip(data_points, labels):
    print(f'Point {point} is in cluster {cluster}')

# Print results for new data points
print("\nNew Data Points and Their Predicted Clusters:")
for point, cluster in zip(new_data_points, predicted_clusters):
    print(f'Point {point} is in cluster {cluster}')

# Print centroids
print("\nCentroids of the Clusters:")
for i, centroid in enumerate(centroids):
    print(f'Centroid {i}: {centroid}')

# Plotting
plt.scatter(data_points[:, 0], data_points[:, 1], c=labels, cmap='viridis', label='Original Data Points')
plt.scatter(new_data_points[:, 0], new_data_points[:, 1], c='red', marker='x', label='New Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='X', s=200, label='Centroids')

plt.title('K-means Clustering with Closely Grouped Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```
