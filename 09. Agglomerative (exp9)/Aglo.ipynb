```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Define the data points
data_points = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [4, 2],
    [4, 4],
    [4, 0],
    [7, 2],
    [7, 4],
    [7, 0]
])

# Set the number of clusters
k = 3

# Create an AgglomerativeClustering instance
agglo_cluster = AgglomerativeClustering(n_clusters=k, linkage='ward')
labels = agglo_cluster.fit_predict(data_points)

# Plotting the clustering results
plt.figure(figsize=(8, 6))
plt.scatter(data_points[:, 0], data_points[:, 1], c=labels, s=100, cmap='viridis', marker='o', label='Data Points')
plt.title('Agglomerative Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
plt.show()

# Plotting the dendrogram
plt.figure(figsize=(10, 7))
linked = linkage(data_points, 'ward')
dendrogram(linked, orientation='top', labels=range(1, len(data_points) + 1), distance_sort='ascending', show_leaf_counts=True)
plt.title('Dendrogram for Agglomerative Clustering')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()

# Print the results
print("Cluster Labels:", labels)

from google.colab import output
output.no_vertical_scroll()
```
