```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

X = np.array([[0, 0], # Input 1
[0, 1], # Input 2
[1, 0], # Input 3
[1, 1]]) # Input 4
y = np.array([0, 1, 1, 1]) # OR outputs
# Build a simple neural network with 1 hidden layer
model = Sequential()
# Input layer (2 neurons for the 2 inputs), and 1 neuron in the output layer (for the bi
model.add(Dense(units=1, input_dim=2, activation='sigmoid')) # Using sigmoid activation
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Train the model
model.fit(X, y, epochs=1000, verbose=0)
# Evaluate the model
loss, accuracy = model.evaluate(X, y)
print(f"Model accuracy: {accuracy*100:.2f}%")
# Make predictions
predictions = model.predict(X)
predictions = (predictions > 0.5).astype(int) # Convert output probabilities to 0 or 1
print("Predictions:")
print(predictions)
```
